```python
import pandas as pd

# Load the dataset
file_path = "C:\\Users\\DeepakArun\\Desktop\\TWC new\\Data\\Marketing App Attribution - Combined - 2023.xlsx"
sheet_name = 'Apps_NewUsers_Visits_Analysis' 
df1 = pd.read_excel(file_path, sheet_name)
```

```python
# Drop the first 3 rows and the last row
df1 = df1.iloc[3:-1].reset_index(drop=True)
```

```python
df1.columns
```

```python
df1
```

```python
# List of columns to subset
columns_to_subset = [
    'Month', 'iOS - MV/MNU', 'Android - MV/MNU'
    
]

# Subsetting the DataFrame
df2 = df1[columns_to_subset]
```

```python
df2
```

```python
df2['iOS - MV/MNU'] = df2['iOS - MV/MNU'].round(2)
df2['Android - MV/MNU'] = df2['Android - MV/MNU'].round(2)
```

```python
import pandas as pd

# Load the dataset
file_path = "C:\\Users\\DeepakArun\\Desktop\\Data\\Data_for_MMM.csv"
df = pd.read_csv(file_path)
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Extract the month from the 'Date' column and add it as a new column
df['Month'] = df['Date'].dt.month

# Merge df1 and df2 based on the 'Month' column
df_merged = pd.merge(df, df2, on='Month', how='left')
```

```python
df_merged
```

```python
df_merged.columns
```

```python
df3 = df_merged.copy()
df3['iOS_New_Users_Computed']= df3['Apps_TWC Universal iOS 4G+_Visits']/df3['iOS - MV/MNU']
df3['Android_New_Users_Computed']= df3['Apps_TWC Universal Android 4G+_Visits']/df3['Android - MV/MNU']
```

```python
df3['App_New_Visits'] = df3['iOS_New_Users_Computed']+df3['Android_New_Users_Computed']
```
