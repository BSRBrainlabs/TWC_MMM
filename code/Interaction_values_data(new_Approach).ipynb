```python
import pandas as pd

# Load the dataset
file_path = "C:\\Users\\DeepakArun\\Desktop\\TWC New\\Data_for_MMM.csv"
df = pd.read_csv(file_path)


# Creating the sum of all paid spend
df['Paid_Media_Spend'] = (
    df['SEM_Google_Spend'] +
    df['Preload_IronSource_Aura_Spend'] +
    df['Programmatic_Bidease_Spend'] +
    df['Preload_Digital_Turbine_Spend'] +
    df['Programmatic_Tapjoy_Spend'] +
    df['SEM_Apple_Search_Ads_Spend'] +
    df['Programmatic_LiftOff_Spend'] +
    df['Programmatic_IronSource_Sonic_Spend'] +
    df['Programmatic_Persona.ly_Spend'] +
    df['Twitter_TikTok_Combined_Spend'] +
    df['Brand_Spend'] +
    df['Influencer_Spend']
)
```

```python
# Creating new columns with the suffix _Paid_Spend by multiplying Organic columns with Paid_Media_Spend
df['OrganicSearch_Google_Clicks_Paid_Spend'] = df['OrganicSearch_Google_Clicks'] * df['Paid_Media_Spend']
df['SocialEng_Facebook_Estimated_Clicks_Paid_Spend'] = df['SocialEng_Facebook_Estimated_Clicks'] * df['Paid_Media_Spend']
df['SocialEng_YouTube_Estimated_Clicks_Paid_Spend'] = df['SocialEng_YouTube_Estimated_Clicks'] * df['Paid_Media_Spend']
df['SocialEng_Instagram_Estimated_Clicks_Paid_Spend'] = df['SocialEng_Instagram_Estimated_Clicks'] * df['Paid_Media_Spend']
df['SocialEng_Twitter_Total_Engagements_Paid_Spend'] = df['SocialEng_Twitter_Total_Engagements'] * df['Paid_Media_Spend']
df['SocialEng_LinkedIn_Impressions_Paid_Spend'] = df['SocialEng_LinkedIn_Impressions'] * df['Paid_Media_Spend']
df['SocialEng_TikTok_Impressions_Paid_Spend'] = df['SocialEng_TikTok_Impressions'] * df['Paid_Media_Spend']
```

```python
# List of columns to subset
columns_to_subset = [
    'Date',
    'Preference',
    'US Weather Impact Rating',
    'Net Trust',
    'NPS',
    'EventsCamp/Vend_Campaign_Flag',
    'Visits_Critical_Event_Flag',
    
    'SEM_Google_Spend',
    'Preload_IronSource_Aura_Spend',
    'Programmatic_Bidease_Spend',
    'Preload_Digital_Turbine_Spend',
    'Programmatic_Tapjoy_Spend',
    'SEM_Apple_Search_Ads_Spend',
    'Programmatic_LiftOff_Spend',
    'Programmatic_IronSource_Sonic_Spend',
    'Programmatic_Persona.ly_Spend',
    'Twitter_TikTok_Combined_Spend',
    'Brand_Spend',
    'Influencer_Spend',
    
    'SEM_Google_Impressions',
    'Preload_IronSource_Aura_impressions',
    'Programmatic_Bidease_Clicks',
    'Preload_Digital_Turbine_launches',
    'Programmatic_Tapjoy_Clicks',
    'SEM_Apple_Search_Ads_Impressions',
    'Programmatic_LiftOff_Impressions',
    'Programmatic_IronSource_Sonic_Impressions',
    'Programmatic_Persona.ly_Clicks',
    'Twitter_TikTok_Combined_Impressions',
    'Brand_Impressions',
    'Influencer_Daily_Impressions',
    
    'OrganicSearch_Google_Clicks',
    'SocialEng_Twitter_Total_Engagements',
    'SocialEng_Facebook_Estimated_Clicks',
    'SocialEng_LinkedIn_Impressions',
    'SocialEng_TikTok_Impressions',
    'SocialEng_YouTube_Estimated_Clicks',
    'SocialEng_Instagram_Estimated_Clicks',
    
    'OrganicSearch_Google_Clicks_Paid_Spend',
    'SocialEng_Facebook_Estimated_Clicks_Paid_Spend',
    'SocialEng_YouTube_Estimated_Clicks_Paid_Spend',
    'SocialEng_Instagram_Estimated_Clicks_Paid_Spend',
    'SocialEng_Twitter_Total_Engagements_Paid_Spend',
    'SocialEng_LinkedIn_Impressions_Paid_Spend',
    'SocialEng_TikTok_Impressions_Paid_Spend',
    'Overall_Web_Visits'
]

# Subsetting the DataFrame
subset_df = df[columns_to_subset]
```

```python
from sklearn.preprocessing import MinMaxScaler

# List of columns to exclude from scaling
exclude_columns = ['Date']

# Columns to scale (all columns except the excluded ones)
columns_to_scale = [col for col in subset_df.columns if col not in exclude_columns]

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply the scaler to the columns that need scaling
subset_df[columns_to_scale] = scaler.fit_transform(subset_df[columns_to_scale])
```
