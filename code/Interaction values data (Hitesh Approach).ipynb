```python
import pandas as pd

# Load the dataset
file_path = "C:\\Users\\DeepakArun\\Desktop\\TWC New\\Latest ADS data (Preprocessing).xlsx"
df = pd.read_excel(file_path)

df = df.fillna(0)

# Assuming df is your DataFrame
df['Paid_Media_Spend'] = (
    df['SEM_Google_Spend'] +
    df['Preload_IronSource_Aura_Spend'] +
    df['Programmatic_Bidease_Spend'] +
    df['Preload_Digital_Turbine_Spend'] +
    df['Programmatic_Tapjoy_Spend'] +
    df['SEM_Apple_Search_Ads_Spend'] +
    df['Programmatic_LiftOff_Spend'] +
    df['Programmatic_IronSource_Sonic_Spend'] +
    df['Programmatic_Persona.ly_Spend'] +
    df['Twitter_TikTok_Combined_Spend'] +
    df['Brand_Spend'] +
    df['Influencer_Spend']
)
```

```python
# Creating new columns with the suffix _Paid_Spend by multiplying with Paid_Media_Spend
df['OrganicSearch_Google_Clicks_Paid_Spend'] = df['OrganicSearch_Google_Clicks'] * df['Paid_Media_Spend']
df['SocialEng_Facebook_Estimated_Clicks_Paid_Spend'] = df['SocialEng_Facebook_Estimated_Clicks'] * df['Paid_Media_Spend']
df['SocialEng_YouTube_Estimated_Clicks_Paid_Spend'] = df['SocialEng_YouTube_Estimated_Clicks'] * df['Paid_Media_Spend']
df['SocialEng_Instagram_Estimated_Clicks_Paid_Spend'] = df['SocialEng_Instagram_Estimated_Clicks'] * df['Paid_Media_Spend']

df['SocialEng_Twitter_Total_Engagements_Paid_Spend'] = df['SocialEng_Twitter_Total_Engagements'] * df['Paid_Media_Spend']

df['SocialEng_LinkedIn_Impressions_Paid_Spend'] = df['SocialEng_LinkedIn_Impressions'] * df['Paid_Media_Spend']
df['SocialEng_TikTok_Impressions_Paid_Spend'] = df['SocialEng_TikTok_Impressions'] * df['Paid_Media_Spend']
```
